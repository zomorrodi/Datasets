```python
import numpy as np

data = np.array(
    [[0.08,0.72,1.0],
     [0.01,1.00,0.0],
     [0.26,0.58,1.0],
     [0.35,0.95,0.0],
     [0.45,0.15,1.0],
     [0.60,0.30,1.0],
     [0.70,0.65,0.0],
     [0.92,0.45,0.0]]
)
w1 = [1.00, -1.00]
w2 = [1.00, -1.00]
arr1 =  np.array([w1,w2])
arr2 =  np.array([w1])
b1 = 0.20
b2 = 0.20

def sigmoid(x):
    return 1.0/ (1.0 + np.exp(-x))
  
def sigmoid_derivative(x):
    sigmoid_fx = sigmoid(x)
    return sigmoid_fx * (1 - sigmoid_fx)    

def loss(predicted, actual):
    return (predicted - actual) ** 2
 
def loss_derivative(predicted, actual):
    return 2 * (predicted - actual)

def acc(data, w1, w2, b1, b2):
    data_x = data[:,:-1]
    data_y = data[:, -1]
    
    tmp1 = sigmoid(np.dot(w1, data_x.T) + b1)
    
    p = np.where(sigmoid(np.dot(w2, tmp1) + b2) >= 0.5, 1.0, 0.0)
    return np.sum(p == data_y) / data.shape[0]

print('\nNOT RANDOM DATA:')

def train(data, w1, w2, b1, b2, epoch_number, rate):
    
    for epoch in range(1, epoch_number+1):
        data_x = data[:,:-1]
        data_y = data[:, -1]
        
        tmp1 = np.dot(w1, data_x.T) + b1
        tmp1_der = sigmoid_derivative(tmp1)
        pred1 = sigmoid(tmp1)
        
        
        tmp2 = np.dot(w2, pred1) + b2
        tmp2_der = sigmoid_derivative(tmp2)
        pred2 = sigmoid(tmp2)

        loss_der = loss_derivative(pred2, data_y)
        update_der = loss_derivative(pred2, data_y) * tmp2_der
        
        w1_update = np.dot(np.dot((update_der).T, w2).T * tmp1_der, data_x) * rate 
        w1 -= w1_update
        w2_update = np.dot(update_der, pred1.T) * rate
        w2 -= w2_update
        b1_update = np.dot((update_der).T,w2).T * np.sum(tmp1_der) * rate
        b1 -= b1_update
        b2_update = np.sum(update_der) * rate
        b2 -= b2_update

        print('\nEpoch:', epoch)
        print('-> Acc =', acc(data, w1, w2, b1, b2))

    return w1, w2, b1, b2

train(data, arr1, arr2, b1, b2, 500, 0.001)
```

```python
#random data
data_x = np.random.randn(8, 2)
data_y = np.random.randint(0, 2, size=(8, 1))
data_rand = np.concatenate((data_x, data_y), axis=1)

w1_rand = np.random.randn(2, 2)
w2_rand = np.random.randn(1, 2)

b1_rand = np.random.randn()
b2_rand = np.random.randn()

print('\nRANDOM DATA:')
train(data_rand, w1_rand, w2_rand, b1_rand, b2_rand, 1000, 0.001)
```
